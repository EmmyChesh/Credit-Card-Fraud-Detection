```python
# Separating The Data For Analysis

legit = df [df.Class == 0]
fraud = df [df.Class == 1]
```

```python
# Compare The Values Of Both Transactions
df.groupby('Class').mean()
```

```python
# Dealing with unbalanced data so we use under sampling
# we will build a sample dataset containing similar distribution of both legit and fraudulent transactions

# Since the number of fraudulent transactions is 492, we will also take 492 random transactions from the legit transaction
# making it a balanced data and will aid in our modeling

legit = legit.sample (n = 492)
legit.shape
```

```python
# now we concatenate the two DataFrames

new_dataset = pd.concat ([legit, fraud], axis = 0)
new_dataset.head()
```

```python
new_dataset.shape
```

```python
# Class count of our new dataset
new_dataset['Class'].value_counts()
```

```python
# It is visible that the nature of the data has not changed, almost similar means are observed and same difference

new_dataset.groupby('Class').mean()
```

```python
# split our features into features and target

x = new_dataset.drop(['Class'], axis = 1)
y = new_dataset ['Class']

x.head()
```

```python
y.head()
```

```python
#split the data into training and testing
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split (x, y, stratify = y, test_size = 0.2, random_state = 2)
```

```python
# check the shapes of our split
print (x.shape, x_train.shape, x_test.shape)
```

## MODEL TRAINING

```python
# import our Logistic Regression Model and training our data
from sklearn.linear_model import LogisticRegression
model = LogisticRegression ()
model.fit (x_train, y_train)
```

```python
# check our model score
train_score = model.score(x_train, y_train)
print ('Accuracy on Training Data:', train_score)
```

```python
# Check our accuracy on our test data

test_score = model.score (x_test, y_test)
print ('Accuracy on Test Data:', test_score)
```

```python
predictions = model.predict (x_test)
predictions
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

score=r2_score(predictions,y_test)
mae= mean_absolute_error(predictions,y_test)
mse= mean_squared_error(predictions,y_test)
rmse=np.sqrt(mse)
```

```python
print (score)
print (mae)
print (mse)
print (rmse)
```
